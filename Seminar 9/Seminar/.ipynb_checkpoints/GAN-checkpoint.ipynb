### Create generator and discriminator

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.model = nn.Sequential( 
            nn.Linear(config.noise_size, 200),
            nn.ReLU(inplace=True),
            nn.Linear(200, config.img_size**2),
            nn.Sigmoid())

    def forward(self, x):
        return self.model(x)
    
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(config.img_size**2, 200),
            nn.ReLU(inplace=True),
            nn.Linear(200, 50),
            nn.ReLU(inplace=True),
            nn.Linear(50, 1), 
            nn.Sigmoid())
        


    def forward(self, x):
        return self.model(x)
```

```python
generator = Generator().cuda()
discriminator = Discriminator().cuda()
```

### Create optimizers and loss

```python
optim_G = optim.Adam(params=generator.parameters(), lr=0.0001)
optim_D = optim.Adam(params=discriminator.parameters(), lr=0.0001)

criterion = nn.BCELoss()
```

# DCGAN

Для DCGAN сделал скрипт на пайтоне и запускал на Ломоносове-2.

```python
class DCgenerator(nn.Module):
    # initializers
    def __init__(self, d=128):
        super(DCgenerator, self).__init__()
        self.deconv1 = nn.ConvTranspose2d(100, d*8, 4, 1, 0)
        self.deconv1_bn = nn.BatchNorm2d(d*8)
        self.deconv2 = nn.ConvTranspose2d(d*8, d*4, 4, 2, 1)
        self.deconv2_bn = nn.BatchNorm2d(d*4)
        self.deconv3 = nn.ConvTranspose2d(d*4, d*2, 4, 2, 1)
        self.deconv3_bn = nn.BatchNorm2d(d*2)
        self.deconv4 = nn.ConvTranspose2d(d*2, d, 4, 2, 1)
        self.deconv4_bn = nn.BatchNorm2d(d)
        self.deconv5 = nn.ConvTranspose2d(d, 1, 4, 2, 1)

    # weight_init
    def weight_init(self, mean, std):
        for m in self._modules:
            normal_init(self._modules[m], mean, std)

    # forward method
    def forward(self, input):
        # x = F.relu(self.deconv1(input))
        x = F.relu(self.deconv1_bn(self.deconv1(input)))
        x = F.relu(self.deconv2_bn(self.deconv2(x)))
        x = F.relu(self.deconv3_bn(self.deconv3(x)))
        x = F.relu(self.deconv4_bn(self.deconv4(x)))
        x = F.tanh(self.deconv5(x))

        return x

class DCdiscriminator(nn.Module):
    # initializers
    def __init__(self, d=128):
        super(DCdiscriminator, self).__init__()
        self.conv1 = nn.Conv2d(1, d, 4, 2, 1)
        self.conv2 = nn.Conv2d(d, d*2, 4, 2, 1)
        self.conv2_bn = nn.BatchNorm2d(d*2)
        self.conv3 = nn.Conv2d(d*2, d*4, 4, 2, 1)
        self.conv3_bn = nn.BatchNorm2d(d*4)
        self.conv4 = nn.Conv2d(d*4, d*8, 4, 2, 1)
        self.conv4_bn = nn.BatchNorm2d(d*8)
        self.conv5 = nn.Conv2d(d*8, 1, 4, 1, 0)

    # weight_init
    def weight_init(self, mean, std):
        for m in self._modules:
            normal_init(self._modules[m], mean, std)

    # forward method
    def forward(self, input):
        x = F.leaky_relu(self.conv1(input), 0.2)
        x = F.leaky_relu(self.conv2_bn(self.conv2(x)), 0.2)
        x = F.leaky_relu(self.conv3_bn(self.conv3(x)), 0.2)
        x = F.leaky_relu(self.conv4_bn(self.conv4(x)), 0.2)
        x = F.sigmoid(self.conv5(x))

        return x

def normal_init(m, mean, std):
    if isinstance(m, nn.ConvTranspose2d) or isinstance(m, nn.Conv2d):
        m.weight.data.normal_(mean, std)
        m.bias.data.zero_()
```

```python
dcG = DCgenerator()
dcD = DCdiscriminator()
dcG.weight_init(mean=0.0, std=0.02)
dcD.weight_init(mean=0.0, std=0.02)
BCE_loss = nn.BCELoss()
G_optimizer = optim.Adam(dcG.parameters(), lr=0.0002, betas=(0.5, 0.999))
D_optimizer = optim.Adam(dcD.parameters(), lr=0.002, betas=(0.5, 0.999))
```

```python
transform = transforms.Compose([
        transforms.Scale(64),
        transforms.ToTensor()])
train = torchvision.datasets.FashionMNIST("fashion_mnist", train=True, transform=transform, download=True)
dataloader = DataLoader(train, batch_size=config.batch_size, shuffle=True, num_workers=config.num_workers)
```

```python
print('training start!')
start_time = time.time()
for epoch in range(config.num_epochs):
    D_losses = []
    G_losses = []
    epoch_start_time = time.time()
    for x_, _ in dataloader:
        # train discriminator D
        dcD.zero_grad()

        mini_batch = x_.size()[0]

        y_real_ = torch.ones(mini_batch)
        y_fake_ = torch.zeros(mini_batch)

        x_, y_real_, y_fake_ = Variable(x_), Variable(y_real_), Variable(y_fake_)
        D_result = dcD(x_).squeeze()
        D_real_loss = BCE_loss(D_result, y_real_)

        z_ = torch.randn((mini_batch, 100)).view(-1, 100, 1, 1)
        z_ = Variable(z_)
        G_result = dcG(z_)

        D_result = dcD(G_result).squeeze()
        D_fake_loss = BCE_loss(D_result, y_fake_)
        D_fake_score = D_result.data.mean()

        D_train_loss = D_real_loss + D_fake_loss

        D_train_loss.backward()
        D_optimizer.step()

        # D_losses.append(D_train_loss.data[0])
        D_losses.append(D_train_loss.item())

        # train generator G
        dcG.zero_grad()

        z_ = torch.randn((mini_batch, 100)).view(-1, 100, 1, 1)
        z_ = Variable(z_)

        G_result = dcG(z_)
        D_result = dcD(G_result).squeeze()
        G_train_loss = BCE_loss(D_result, y_real_)
        G_train_loss.backward()
        G_optimizer.step()

        G_losses.append(G_train_loss.item())

    epoch_end_time = time.time()
    per_epoch_ptime = epoch_end_time - epoch_start_time


    print('[%d/%d] - ptime: %.2f, loss_d: %.3f, loss_g: %.3f' % ((epoch + 1), config.num_epochs, per_epoch_ptime, torch.mean(torch.FloatTensor(D_losses)),
                                                              torch.mean(torch.FloatTensor(G_losses))))
```

# cGAN

```python
class cGenerator(nn.Module): 
    def __init__(self):
        super(cGenerator, self).__init__()
        self.label_embed = nn.Embedding(config.num_labels, config.num_labels)
        self.depth=128

        def init(input, output, normalize=True): 
            layers = [nn.Linear(input, output)]
            if normalize: 
                layers.append(nn.BatchNorm1d(output, 0.8))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return layers 

        self.generator = nn.Sequential(

            *init(config.noise_size+config.num_labels, self.depth), 
            *init(self.depth, self.depth*2), 
            *init(self.depth*2, self.depth*4), 
            *init(self.depth*4, self.depth*8),
            nn.Linear(self.depth * 8, config.img_size**2),
            nn.Tanh()

            )

    # torchcat needs to combine tensors 
    def forward(self, noise, labels): 
        gen_input = torch.cat((self.label_embed(labels), noise), -1)
        img = self.generator(gen_input)
        img = img.view(img.size(0), config.channels, config.img_size, config.img_size)
        return img


class cDiscriminator(nn.Module): 
    def __init__(self): 
        super(cDiscriminator, self).__init__()
        self.label_embed1 = nn.Embedding(config.num_labels, config.num_labels)
        self.dropout = 0.4
        self.depth = 512

        def init(input, output, normalize=True): 
            layers = [nn.Linear(input, output)]
            if normalize: 
                layers.append(nn.Dropout(self.dropout))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return layers 

        self.discriminator = nn.Sequential(
            *init(config.num_labels+ config.img_size**2, self.depth, normalize=False),
            *init(self.depth, self.depth), 
            *init(self.depth, self.depth),
            nn.Linear(self.depth, 1),
            nn.Sigmoid()
            )

    def forward(self, img, labels): 
        imgs = img.view(img.size(0),-1)
        inpu = torch.cat((imgs, self.label_embed1(labels)), -1)
        validity = self.discriminator(inpu)
        return validity 

def init_weights(m): 
    if type(m)==nn.Linear:
        torch.nn.init.xavier_uniform(m.weight)
        m.bias.data.fill_(0.01)
```

```python
G = cGenerator().cuda()
D = cDiscriminator().cuda()
optim_G = optim.Adam(G.parameters(), lr=0.0002,betas=(0.5, 0.999))
optim_D = optim.Adam(D.parameters(), lr=0.0002,betas=(0.5, 0.999))
D.apply(init_weights)
BCE_loss = nn.BCELoss().cuda()
real_label = 0.9
fake_label = 0.0
FT = torch.cuda.LongTensor
FT_a = torch.cuda.FloatTensor
```

```python
print('training start!')
start_time = time.time()
for epoch in range(config.num_epochs):
    epoch_start_time = time.time()
    for x_, y_ in dataloader:
        x_ = Variable(x_.type(FT_a))
        y_ = Variable(y_.type(FT))
        # train discriminator D
        optim_D.zero_grad()
        optim_G.zero_grad()
        mini_batch = x_.size(0)
        
        y_real = Variable(FT_a(mini_batch,1).fill_(real_label))
        y_fake = Variable(FT_a(mini_batch,1).fill_(fake_label))
        
        noise = Variable(FT_a(np.random.normal(0, 1,(mini_batch, config.noise_size))))
        gen_y = Variable(FT(np.random.randint(0, config.num_labels, mini_batch)))
        
        gen_imgs = G(noise, gen_y)
        
        D_result = D(gen_imgs, gen_y)
        G_loss = BCE_loss(D_result, y_real)
        
        G_loss.backward()
        optim_G.step()
        
        
        optim_D.zero_grad()
        
        D_real=D(x_, y_)
        D_real_loss=BCE_loss(D_real, y_real)
        
        D_fake=D(gen_imgs.detach(), gen_y)
        D_fake_loss=BCE_loss(D_fake, y_fake)
        
        D_loss=0.5*(D_fake_loss+D_real_loss)
        D_loss.backward()
        optim_D.step()
        

    epoch_end_time = time.time()
    per_epoch_ptime = epoch_end_time - epoch_start_time
    print('[%d/%d] - ptime: %.2f, loss_d: %.3f, loss_g: %.3f' % ((epoch + 1), config.num_epochs, per_epoch_ptime, D_loss.item(), G_loss.item()))
```
