## Задание 1

Напишите свою реализацию LSTM.

```python
class LongShort(nn.Module):
    def __init__(self,inp_size,mem_size):
        super(LongShort, self).__init__()
        self.inp_size=inp_size
        self.mem_size=mem_size
        self.i2h=nn.Linear(inp_size,4*mem_size)
        self.h2h=nn.Linear(mem_size,4*mem_size)
    
    def forward(self,inputs):
        preact=self.h2h(inputs[1])+self.i2h(inputs[2])
        forget_gate=nn.Sigmoid()((preact).narrow(2,0,self.mem_size))
        input_gate_sigm=nn.Sigmoid()((preact).narrow(2,self.mem_size,self.mem_size))
        input_gate_tanh=nn.Tanh()((preact).narrow(2,2*self.mem_size,self.mem_size))
        output_gate=nn.Sigmoid()((preact).narrow(2,3*self.mem_size,self.mem_size))
        c_forget=inputs[0]*forget_gate
        c_transformed=c_forget+(input_gate_sigm*input_gate_tanh)
        result=nn.Tanh()(c_transformed)*output_gate
        return {result,c_transformed}
```

Данные о количестве пассажиров
https://datamarket.com/data/set/22u3/international-airline-passengers-monthly-totals-in-thousands-jan-49-dec-60#!ds=22u3&display=line

```python
%matplotlib inline
dataset = pandas.read_csv('international-airline-passengers.csv', usecols=[1], delimiter=";", engine='python', skipfooter=3)
plt.plot(dataset)
plt.show()
```

```python
# подготовливаем данные

dataset = dataset.values
dataset = dataset.astype('float32')
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
```

```python
# разбиваем данные на train / test
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
print(len(train), len(test))
```

```python
# Целевая переменная - сдвинутая последовательность значений входов

def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])

    return np.array(dataX), np.array(dataY)
```

```python
# Пример подготовленных данных

look_back = 10
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
```

```python
NN=nn.LSTM(10,10,1)
X=Variable(torch.Tensor(trainX))
h0=Variable(torch.zeros(1,1,10))
c0=Variable(torch.zeros(1,1,10))
loss_fn=nn.L1Loss()
optimizer = torch.optim.Adam(NN.parameters(), lr=0.04)
old_h=Variable(torch.zeros(1,1,10))
old_c=Variable(torch.zeros(1,1,10))
```

```python
for epoch in xrange(1):
    for i in xrange(trainX.shape[0]):
        x=Variable(torch.Tensor(trainX[i]))
        y_check=Variable(torch.FloatTensor([trainY[i].astype(float)]),requires_grad=False)
        (new_h,new_c)=NN.forward(x,(old_h,old_c))[1]
        y=new_h.narrow(2,9,9)
        loss=loss_fn(y,y_check)
        optimizer.zero_grad()
        loss.backward(retain_graph=True)
        optimizer.step()
        (old_h,old_c)=(new_h,new_c)
```

```python
x=Variable(torch.Tensor(trainX[0]),requires_grad=False)
newy=np.zeros(len(trainY))
result=np.zeros(len(trainY))
for i in xrange(trainX.shape[0]):
        x=Variable(torch.Tensor(trainX[i]))
        y_check=Variable(torch.FloatTensor([trainY[i].astype(float)]),requires_grad=False)
        (new_h,new_c)=NN.forward(x,(old_h,old_c))[1]
        newy[i]=new_h.narrow(2,9,9).data[0][0][0]
        (old_h,old_c)=(new_h,new_c)
        result[i]=trainX[i][0][0]
plt.figure()
plt.plot(result)
plt.plot(trainY)
```
